# Modelo de programación matemática para el Sudoku


Para resolver el Sudoku es posible emplear un modelo de programación lineal entera.

Las variables de decisión se definen de la siguiente manera:

$x_{ijk} = 1$, si la celda (*i*, *j*) del Sudoku contiene el número *k*.

$x_{ijk} = 0$, en otro caso.

Donde $1 \leq i,j,k \leq 9$.

#Función objetivo
En este caso no se tiene función objetivo, ya que el Sudoku se resuelve en cuanto se satisfacen todas las restricciones.

Algunos autores ponen como función objetivo minimizar una constante, por ejemplo:

$Min$ $f(x)=0$

#Restricciones

El modelo tiene un conjunto de ecuaciones que fuerzan la asignación de exactamente un número del 1 al 9 a cada celda:

$\sum_{k=1}^{9}x_{ijk}=1$, para toda $1 \leq i,j \leq 9$

Otra familia de restricciones representa la ocurrencia única de cada valor en cada columna:

$\sum_{i=1}^{9}x_{ijk}=1$, para toda $1 \leq n,k \leq 9$

Para garantizar que cada número aparezca una vez en cada fila:

$\sum_{j=1}^{9}x_{ijk}=1$, para toda $1 \leq i,k \leq 9$

Para que en cada sub-cuadrado aparezca una vez dada número se tiene:

$\sum_{i=h_0}^{h_0+2}\sum_{j=h_1}^{h_1+2}x_{ijk}=1$, para toda $1 \leq h \leq 9$,  $h_0,h_1 \in \{1, 4, 7 \}$

También se define el conjunto *S* con las celdas que vienen llenadas desde el inicio del Sudoku, y se emplean la siguiente familia de restricciones.

$x_{ijk}=1$,  $∀(i,j) \in S$

Finalmente se agrega el tipo de variables empleadas.

$x_{ijk} \in \{0,1\}$
